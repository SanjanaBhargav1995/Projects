```python
df_train["Title"]=df_train['Name'].str.extract(" ([A-Za-z]+)\.")
df_train.head()
```

```python
df_train_cat = df_train.select_dtypes(exclude=np.number)
df_train_num = df_train.select_dtypes(include=np.number)
```

```python
df_train_cat
```

```python
df_train["Title"].value_counts()
```

```python
df_train_num
```

```python
df_train_cat["Title"]= df_train_cat["Title"].str.strip()

df_train_cat["Title"] = np.where((df_train_cat["Title"] != "Mr")&
                        (df_train_cat["Title"] != "Miss") &
                        (df_train_cat["Title"] != "Mrs") &
                        (df_train_cat["Title"] != "Master"),
                        "others",df_train["Title"])
```

```python
df_train_cat.drop(["Name", "Ticket"], axis=1,inplace=True)
```

```python
df_train_cat
```

```python
df_train_cat["is_adult"] = df_train_cat["is_adult"].map({
    True : 1,
    False : 0
})
```

```python
df_train_cat = pd.get_dummies(df_train_cat,drop_first=True)
# df_train_cat = pd.get_dummies(df_train_cat,drop_first=True,dtype="int")---To convert data in binary
```

```python
df_train_cat
```

```python
# Making copies 1- Deep copy 2 Shallow copy
df_test = df_train_cat.copy()
df_train.rename(columns = lambda x: x.lower())
```

```python
# merging df_train_cat and df_train_num
# merge should have unique value to join
# concanate method
df_final =pd.concat([df_train_cat,df_train_num],axis =1)
df_final.head()
```

```python
df_final.to_csv("titanic_cleaned.csv",index=False)
```

```python
X = df_final.drop(["Survived"], axis = 1)
y = df_final[["Survived"]]
```

```python
df_final
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size =0.33, random_state=42)
```

```python
lr = LogisticRegression(class_weight = "balanced")
lr.fit(X_train, y_train)

y_pred = lr.predict(X_test)
y_pred_prob = lr.predict_proba(X_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```
